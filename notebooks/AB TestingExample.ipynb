## Normal Distribution

```python
N_A = 1000  # number of impressions for control group
N_B = 1000  # number of impressions for test group
X_A = 815  # clicks for control group
X_B = 815  # clicks for test group
```

```python
bcr = X_A / N_A  # proportion of clicks over impressions for control group
d_hat = X_B / N_B - bcr  # difference in click proportions between test and control
bcr, d_hat
```

```python
abplot(N_A, N_B, bcr, d_hat, show_power=False, show_alpha=True)
```

## Significance Level and Confidence Level

* When N is 100, the 95% confidence intervals are at -0.11 to 0.11
* When N is 1000, the 95% confidence intervals are at -0.04 to 0.04
* When our sample size increases, our 95% confidence interval shrinks.
* Every time you assess a random sample of 1000 impressions, 95% of the time the proportion will be between 0.811 to 0.819.
